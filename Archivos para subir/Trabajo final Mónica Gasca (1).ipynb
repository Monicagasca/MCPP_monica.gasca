### <font color="green"> Utilizamos la librería Beautifulsoup y métodos de web scraping para obtener los url de los documentos y resoluciones que tratan sobre el cargo por confiabilidad

### <font color="orange"> Circulares

```python
url = open ("url_circulares.txt")

for row in url: 
    try:
        base= row
        index= urlopen(base)
        soup = BeautifulSoup(index, "lxml")
        circular = soup.find_all("a", href=re.compile("/Publicac.nsf/52188526a7290f8505256eee0072eba7/\d+"))
    except:   
        print("error")
    print (circular)
```

### <font color="orange"> Resoluciones

```python
url_2 = open ("url_resoluciones.txt")

for row in url_2: 
    try:
        base= row
        index= urlopen(base)
        soup = BeautifulSoup(index, "lxml")
        resoluciones = soup.find_all( "a", href=re.compile("/Publicac.nsf/1c09d18d2d5ffb5b05256eee00709c02/\d+"))
    except:   
        print("error")
    print (resoluciones)
```

```python
url_3= open( "url_linksinteres.txt")

for row in url_3: 
        base= row
        index= urlopen(base)
        soup = BeautifulSoup(index, "lxml")
        [x.extract() for x in soup.findAll('head')]
        [x.extract() for x in soup.findAll('font')]
        [x.extract() for x in soup.findAll('br')]
        [x.extract() for x in soup.findAll('div')]
        
        limpio= soup.find_all("a", href=re.compile("/Publicac.nsf/\d+"))
        print (limpio)
```
